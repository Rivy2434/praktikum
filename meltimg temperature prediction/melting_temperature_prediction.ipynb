### Description
    

> Following project is related to the prediction of the steel melting temperature during the production. </br></br>
> In a very simplified way, steel production from raw iron could be described as follows: </br>
Unwanted impurities such as sulfur are removed. Melted steel is supplied into the fireproof pot of 100-tonne capacity. It is heated by graphite electrodes until the necessary temperature. The chemical compound is checked, then different ligating additives are added. The mixture is melted and ventilated with a rare gas, and the temperature is checked. These measurements iterations are repeated until achieving the necessary chemical compound and optimal melting temperature. </br></br>
> Data description:</br>
> - data comprised of different files from different sources</br>
> - key - batch number in every file</br>
> - data_arc - data about electrodes</br>
> - data_bulk - volume of the bulky materials added to the mixture</br>
> - data_bulk_time - time of supply</br>
> - data_gas - data about gas (no additional information)</br>
> - data_temp - temperature measurements results</br>
> - data_wire - volume of added wire materials</br>
> - data_wire_time - time of supply</br>

```python
#function for histogramms creation for every column of the table
def many_hists(data,rows,cols):
    fig,axs = plt.subplots(rows,cols,figsize=(15,18))
    names = data.columns
    for i in range (0,rows):
        for j in range (0,cols): 
            #убираем нулевые значения  - это те, которые были NaN
            axs[i,j].hist(data[data[names[(i)*3+(j)]]!=0][names[(i)*3+(j)]],bins=100) 
            axs[i,j].set_title(names[(i)*3+(j)]) 
    plt.show()   
```

### Model training

#### Additional functions

```python
#automatization of training models
#features_list = [features_train,target_train,features_test,target_test]
def ModelTraining(model,parameters,features_list):
    #making pipeline for further cross-validation
    model_pipe = make_pipeline(StandardScaler(),model)
        
    gsearch_model = GridSearchCV(model_pipe,parameters,scoring='neg_mean_absolute_error',n_jobs=-1,cv=10)
    gsearch_model.fit(features_list[0],features_list[1])
    print('model parameters: ',gsearch_model.best_params_)

    #testing on train and test
    
    
    predictions_train = gsearch_model.best_estimator_.predict(features_list[0])
    predictions_test = gsearch_model.best_estimator_.predict(features_list[2])
    
    MAE_train = mean_absolute_error(features_list[1],predictions_train)
    MAE_test = mean_absolute_error(features_list[3],predictions_test)
    
    #model = model.set_params(gsearch_model.best_params_).fit(features_list[0])
    
    print('trained model:')
    print('train accuracy: {}'.format(MAE_train))
    print('test accuracy: {}'.format(MAE_test))
    return gsearch_model.best_estimator_
```

```python
#accessory function for feature importance bar charts

def importance_plot(model,features_list,nametag): #model name, start and end numbers of features in descending order
    #from feature_importances for model we make Series,where indexes = features names
    model_importance = pd.Series(model.feature_importances_,index=features_list[0].columns) 
    #sorting descending, slicing by start-stop numbers
    model_imp = model_importance.sort_values(ascending=False)
    fig, ax = plt.subplots(figsize=(8,8))
    model_imp.plot.bar()
    title = "Feature importances for " + nametag
    ax.set_title(title)
    fig.tight_layout()
```
